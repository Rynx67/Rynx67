```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
result = classifier(
    "The patient was prescribed a new medication for their chronic pain.",
    candidate_labels=["medical treatment", "symptoms", "diagnosis"]
)
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")
generated_text = generator(
    "As your doctor, I would recommend that you consider...",
    max_length=30,
    num_return_sequences=2
)
for i, text in enumerate(generated_text):
    print(f"Generated Text {i+1}: {text['generated_text']}")
```

```python
from transformers import pipeline

for i, text in enumerate(generated_text):
    print(f"Generated Text {i+1}: {text['generated_text']}")

ner = pipeline("ner", grouped_entities=True)
ner_results = ner("The patient, Kevin Chandra, was diagnosed with Type 2 Diabetes and prescribed Metformin.")
print("NER Results:")
```

```python
from transformers import pipeline

for entity in ner_results:
    print(entity)

sentiment_analysis = pipeline("sentiment-analysis")
feedback = "The treatment was highly effective, and I feel much better now."
sentiment_result = sentiment_analysis(feedback)
print(f"Sentiment Analysis: {sentiment_result}")
```

```python
from transformers import pipeline

summarizer = pipeline("summarization")
research_text = (
'''
    Recent studies have shown that a high-fiber diet can help reduce the risk of heart disease.
    Additionally, regular exercise and maintaining a balanced diet are essential for overall health and well-being.
    Physicians recommend at least 30 minutes of exercise daily along with a diet rich in fruits, vegetables, and whole grains.
'''
)
```

```python
from transformers import pipeline

summary = summarizer(research_text)
print("Summary:")
print(summary[0]['summary_text'])

translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-id")
text_to_translate = "Please take this medicine twice daily after meals."
translated_text = translator(text_to_translate)
print("Translation:")
print(translated_text[0]['translation_text'])
```
